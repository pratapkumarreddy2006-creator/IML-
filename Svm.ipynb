```python
import numpy as np
from sklearn import svm
x=np.array([[2,2],[4,5],[7,4]])
y=np.array([-1,1,1])
clf=svm.SVC(kernel='linear',C=1000)
clf.fit(x,y)
weight=clf.coef_[0]
intercept=clf.intercept_[0]
sUPPORT_vECTORS =clf.support_vectors_
print("--SVM numerical Results ---")
print(f"weights (W): {weight}")
print(f"intercept (b): {intercept}")
print(f"support_vectors: {support_vectors}")
print(f"\n decision boundary equation: {weight[0]:.2f}*x_0 + {weight[1]:.2f}*x_1 + {intercept:.2f} = 0")
```
